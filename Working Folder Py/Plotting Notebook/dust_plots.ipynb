```python
import scipy.io as spio
from matplotlib import pyplot as plt 
import numpy as np 
from pylab import *
```

```python
mat = spio.loadmat('/Users/Isabel/dust_data_large_O2_neg5V.mat', squeeze_me=True)

x = mat['x']
y = mat['y']
vx = mat['vx']
vy = mat['vy']
V = mat['phi']
dust = mat['object1']
rho = mat['den']
Ex = mat['efx']
Ey = mat['efy']
q = mat['chg']

Lx = 0.0015

#print(q)
#v = np.concatenate(vx, vy)
#print(dust)
```

```python
plt.scatter(x,y, 10, 'm')
plt.xlim(0, 0.0015) 
plt.ylim(0, 0.0015)
plt.title('Position: x vs y')

plt.tight_layout()
plt.show()
```

```python
plt.scatter(y, vy, 10, 'm')
plt.xlim(0, 0.0015)
plt.xticks('Lx')
plt.title('Phase plot (y)')
plt.tight_layout()
plt.show()
```

```python
plt.scatter(x,vx, 10, 'm')
plt.xlim(0, 0.0015) 
plt.title('Phase plot in x')
plt.tight_layout()
plt.show()
```

```python
plt.plot(Ex)
plt.xlim(0, 9) 
plt.title('Ex vs x')
plt.tight_layout()
plt.show()
```

```python
plt.plot(V)
plt.xlim(0, 9) 
plt.title('Electric potential')
plt.xlabel('x')
plt.ylabel('V')
plt.tight_layout()
plt.show()
```

```python
plt.plot(rho)
ax.set_prop_cycle(cycler('color', ['c', 'm', 'y', 'k']) )
plt.title('Ion density')

plt.tight_layout()

plt.show()
```

```python
plt.contourf(np.transpose(V),cmap=plt.cm.get_cmap('cool'))
plt.colorbar()
VC = plt.contour(np.transpose(V), colors='k',linewidths= 1)
plt.contourf(dust, [0.9, 1 ], colors ='k')
#manual_locations = [(0.5, 4.5), (1.5 , 4.5), (2, 4.5), (2.5, 4.5)]
#plt.clabel(VC, inline=1, fontsize=13, manual=manual_locations, , colors='k')
plt.title('Electric potential')
plt.xlabel('x')
plt.ylabel('y')
plt.tight_layout()
plt.show()
```

```python
plt.contourf(np.transpose(rho), cmap=plt.cm.get_cmap('plasma') )
plt.colorbar()
plt.contour(np.transpose(rho), colors='k', linewidths= 1)
plt.contourf(dust, [0.9, 1 ], colors ='k')
plt.tight_layout()
plt.title('Ion density')
plt.show()
```
