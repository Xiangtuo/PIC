```python
q = np.zeros((170, 170))

sw = 3683.264706
q_mp = 1
N = 1000
p_pos = np.zeros([N,2])
p_velo = np.zeros([N,2])
dh = 0.006

for p in range (1, N):
    fi = (1 + p_pos[p,0]) / (dh) 
    i = np.floor(fi)
    hx = fi - i
    
    fj = (1 + p_pos[p,1]) / (dh)
    j = np.floor(fj)
    hy = fj - j
    
    q[i,j] = q[i,j] + (1-hx) * (1-hy)
    
rho = (sw + q_mp * q) / (dh * dh)
rho = rho + 1 * 10 ** 4

print(rho)
```
